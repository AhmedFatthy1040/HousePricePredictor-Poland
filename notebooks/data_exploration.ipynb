# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
file_path = '../data/Houses.csv'
try:
    dataset = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    dataset = pd.read_csv(file_path, encoding='latin1')
```

# Visualize the distribution of the target variable 'price'

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['price'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price (PLN)')
plt.ylabel('Frequency')
plt.show()
```

# Visualize the relationship between numerical features and the target variable

```python
sns.pairplot(dataset[['floor', 'rooms', 'sq', 'year', 'price']])
plt.suptitle('Pairplot of Numerical Features')
plt.show()
```

# Visualize the relationship between categorical features and the target variable

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='city', y='price', data=dataset)
plt.title('Boxplot of House Prices by City')
plt.xlabel('City')
plt.ylabel('Price (PLN)')
plt.show()
```
